```python
# 데이터 수집
from sklearn import datasets
X, y = datasets.fetch_openml('boston',return_X_y = True)
# fetch_openml -> fetch_open API가 자동으로 dataFrame으로 변환을 시켜주는 기능
# boston_data = datasets.fetch_openml('boston')
```

### 컬럼정보
- CRIM : 범죄율
- ZN : 주거지 비율 (25000 평방피트 이상의 주거지 비율)
- INDUS : 상업용 지역 비율
- CHAS : 찰스 강의 접근성
- NOX : 대기 중 산화질소 농도
- RM : 평균 방의 개수
- AGE : 1940년 이전에 지어진 주택의 비율
- DIS : 직업센터까지 거리
- RAD : 고속도로 접근성 [클수록 용의함]
- TAX : 재산세율
- PTRATIO : 교사와 학생 비율
- B : 흑인 비율
- LSTAT : 하위 계층 비율

```python
# 데이터 확인
X.info()
```

```python
X[ 'CHAS' ] = X[ 'CHAS' ].astype( 'int64' )
X[ 'RAD' ] = X[ 'RAD' ].astype( 'int64' )
```

### 모델 선택 및 하이퍼 파라미터 튜닝

```python
# 모델 도구 불러오기
from sklearn.linear_model import LinearRegression
# 모델 객체 생성
linear_model = LinearRegression()
```

```python
# 학습용 데이터와 테스트용 데이터를 나누기
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 5)
```

```python
# 데이터를 나누면 항상 해야하는 것
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# 모델 학습
linear_model.fit(X_train, y_train)
```

```python
# 경사하강법 사용
# 경사하강법 사용 도구 불러오기
from sklearn.linear_model import SGDRegressor
# 모델 객체 생성
sgd_model = SGDRegressor(eta0 = 0.1, verbose = 2)
# eta0 = 학습률 -> learning rate 기본값으로는 0.01
# 0 : 출력없음 / 1 : 에포크마다 진행상태 출력 / 2 : 상세한 출력
# 에포크 -> 1회 학습을 의미
```

```python
# 모델 학습
sgd_model.fit(X_train, y_train)
```

```python
from sklearn.preprocessing import LabelEncoder

for col in ['CHAS', 'RAD']:
    le = LabelEncoder()
    X_test[col] = le.fit_transform(X_test[col])
```

```python
# 모델 평가
sgd_model.score(X_test, y_test)
```
